This file is used to create plots for the historical homelessness data in "HomelessnessData.csv"

```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

plt.rcParams.update({
    'figure.figsize':(11,7),
    'figure.dpi':120
})

image_dest = 'results_images/'
plt.rc('font', size=15)
palette = 'Greys'
linewidth = 2
color = 'black'
alpha = 0.2

df = pd.read_csv("HomelessnessData.csv")
old_cols = df.columns.tolist()
old_cols.remove('Year')
df['Spokane-Total (100s)'] = df['Spokane-Total'] / 100
df['Spokane-Unsheltered (100s)'] = df['Spokane-Unsheltered'] / 100
df['WA-Total (1,000s)'] = df['WA-Total'] / 1000
df['WA-Unsheltered (1,000s)'] = df['WA-Unsheltered'] / 1000
df['USA-Total (100,000s)'] = df['USA-Total'] / 100000
df['USA-Unsheltered (100,000s)'] = df['USA-Unsheltered'] / 100000
df = df.drop(old_cols, axis=1)
```

```python
columns = df.columns.tolist()
columns.remove('Year')
line_types = [':', ':', '-.', '-.', '--', '--']
colors = ['black', 'gray']*3
for i in range(len(columns)):
    p = sns.lineplot(data=df, x='Year', y=columns[i], label=columns[i], color=colors[i])
    p.lines[i].set_linestyle(line_types[i])
plt.ylim((0, 40))
plt.ylabel('Number of People')
plt.savefig(fname='Fig1.png', bbox_inches='tight')
plt.show()
```
